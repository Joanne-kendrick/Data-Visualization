```python
#function to assign BMI category
def bmi_category(value):
  if value < 18.5: #Underweight
      return 1
  elif value>=18.5 and value<=24.9: #Normal weight
      return 2
  elif value>=25.0 and value<=29.9: #Overweight
      return 3
  elif value>=30.0 and value<=39.9: # obesity
      return 4
  elif value>=40.0: #obesity
      return 5

df['BMICategory'] = df['BMI'].apply(bmi_category)
```

```python
# convert this new column into category type
order = df.BMICategory.value_counts().index.sort_values()
ordered_var = pd.api.types.CategoricalDtype(ordered=True, categories = order)
df.BMICategory = df.BMICategory.astype(ordered_var)
```

```python
def get_prop(sex, diab_class):
    """This function calculates the proportion of a particular sex
    that falls into a given diabetes classification"""

    return round((len(df.Sex[(df.Sex==sex) & (df.Diabetes_binary==diab_class)]) / \
    len(df.Sex[df.Sex==sex])) * 100, 2)
```

```python
# build two dictionaries to hold each proportions for each gender.
# This is necessary because of the visualization that will follow

female_props_dict = {}
male_props_dict = {}

prefixes = ['no_', 'pre_', '']
genders = ['female', 'male']
for i in range(3):  # loop runs three times, once for each diabetes class
    for j in range(2):  # loop runs twice, once for each sex
        if j==0:
            female_props_dict[f'{prefixes[i]}diab_{genders[j]}_prop'] = get_prop(j, i)
        else:
            male_props_dict[f'{prefixes[i]}diab_{genders[j]}_prop'] = get_prop(j, i)



print(male_props_dict)
print(female_props_dict)
```

```python
y=np.arange(len(female_props_dict))
plt.barh(y=y-0.2, height=0.4, width=male_props_dict.values())
plt.barh(y=y+0.2, height=0.4, width=female_props_dict.values())
plt.yticks(y, ['no diabetes', 'prediabetes', 'diabetes']);
plt.legend(['males', 'females'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import logit

# Set a colorblind-friendly palette and style
sns.set_palette("colorblind")
sns.set_style('whitegrid')

# --- Cleveland Dot Plot: Odds Ratios for Chronic Conditions ---
conditions = ['Stroke', 'HeartDiseaseorAttack', 'HighBP', 'HighChol']
odds_ratios = []
for condition in conditions:
    model = logit(f"Diabetes_binary ~ {condition}", data=df).fit(disp=0)
    odds_ratio = np.exp(model.params[1])
    ci_lower, ci_upper = np.exp(model.conf_int().iloc[1])
    odds_ratios.append((condition, odds_ratio, ci_lower, ci_upper))

# Define a color palette
colors = sns.color_palette("colorblind", len(conditions))

# Create the Cleveland Dot Plot
fig, ax = plt.subplots(figsize=(10, 7))
for i, (condition, or_, ci_lower, ci_upper) in enumerate(odds_ratios):
    ax.plot(or_, i, 'o', color=colors[i], markersize=10, label=condition)
    ax.hlines(i, ci_lower, ci_upper, color=colors[i], linewidth=2)
    ax.text(or_ + 0.1, i, f"{or_:.2f}", va='center', ha='left', fontsize=10, color=colors[i])

ax.set_yticks(range(len(conditions)))
ax.set_yticklabels(['Stroke', 'Heart Disease', 'High BP', 'High Cholesterol'], fontsize=12)
ax.set_xlabel('Odds Ratio', fontsize=14)
ax.set_title('Odds Ratios of Chronic Conditions for Diabetes', fontsize=16, fontweight='bold')
ax.axvline(1, color='gray', linestyle='--', linewidth=1.5)
ax.grid(True, linestyle='--', linewidth=0.5, color='lightgray')
ax.set_xlim(0, max(or_ + (ci_upper - or_) for _, or_, _, ci_upper in odds_ratios) * 1.2)
ax.legend(loc='lower right', fontsize=10)
plt.tight_layout()
plt.savefig('cleveland_dot_odds_ratios.png', dpi=300)
plt.close()

# --- Grouped Line Plot: Diabetes Prevalence by Age, BMI, and Physical Activity ---
# Precompute prevalence data for efficiency
prevalence_data = df.groupby(['PhysActivity', 'Age', 'BMICategory'])['Diabetes_binary'].mean().reset_index()

# Create the grouped line plot
g = sns.relplot(data=prevalence_data, x='Age', y='Diabetes_binary', hue='BMICategory', col='PhysActivity',
                kind='line', marker='o', palette='Blues', height=6, aspect=1.5)

# Customize titles and labels
g.set_titles('Physical Activity: {col_name}', size=14)
g.set_axis_labels('Age Group', 'Diabetes Prevalence', size=12)
g.add_legend(title='BMI Category')
g.fig.suptitle('Diabetes Prevalence by Age and BMI, Faceted by Physical Activity', y=1.02, fontsize=18, fontweight='bold')

# Ensure all age groups are shown
g.set(xticks=range(1, 14))

plt.savefig('grouped_line_prevalence.png', dpi=300)
plt.close()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import logit


# Set a colorblind-friendly palette and style with larger fonts
sns.set_palette("colorblind")
sns.set_style('whitegrid')
sns.set_context("notebook", font_scale=1.5)

# --- Cleveland Dot Plot: Odds Ratios for Chronic Conditions ---
conditions = ['Stroke', 'HeartDiseaseorAttack', 'HighBP', 'HighChol']
odds_ratios = []
for condition in conditions:
    model = logit(f"Diabetes_binary ~ {condition}", data=df).fit(disp=0)
    odds_ratio = np.exp(model.params[1])
    ci_lower, ci_upper = np.exp(model.conf_int().iloc[1])
    odds_ratios.append((condition, odds_ratio, ci_lower, ci_upper))

# Define a color palette
colors = sns.color_palette("Set2", len(conditions))

# Create the Cleveland Dot Plot
fig, ax = plt.subplots(figsize=(10, 7))
for i, (condition, or_, ci_lower, ci_upper) in enumerate(odds_ratios):
    ax.plot(or_, i, 'o', color=colors[i], markersize=10, label=condition)
    ax.hlines(i, ci_lower, ci_upper, color=colors[i], linewidth=2)
    ax.text(or_ + 0.1, i, f"{or_:.2f}", va='center', ha='left', fontsize=10, color=colors[i])

ax.set_yticks(range(len(conditions)))
ax.set_yticklabels(['Stroke', 'Heart Disease', 'High BP', 'High Cholesterol'], fontsize=12)
ax.set_xlabel('Odds Ratio', fontsize=14)
ax.set_title('Odds Ratios of Chronic Conditions for Diabetes', fontsize=16, fontweight='bold')
ax.axvline(1, color='gray', linestyle='--', linewidth=1.5)
ax.grid(True, linestyle='--', linewidth=0.5, color='lightgray')
ax.set_xlim(0, max(or_ + (ci_upper - or_) for _, or_, _, ci_upper in odds_ratios) * 1.2)
ax.legend(loc='lower right', fontsize=10)
ax.text(0.5, -0.1, "Error bars represent 95% confidence intervals.",
        horizontalalignment='center', verticalalignment='center',
        transform=ax.transAxes, fontsize=10)
plt.tight_layout()
plt.savefig('cleveland_dot_odds_ratios.png', dpi=300)
plt.close()

# --- Separate Line Plots: Diabetes Prevalence by Age and BMI for each Physical Activity level ---
# Precompute prevalence data for efficiency
prevalence_data = df.groupby(['PhysActivity', 'Age', 'BMICategory'])['Diabetes_binary'].mean().reset_index()

# Define mappings for age and BMI categories
age_labels = {1: '18-24', 2: '25-29', 3: '30-34', 4: '35-39', 5: '40-44',
              6: '45-49', 7: '50-54', 8: '55-59', 9: '60-64', 10: '65-69',
              11: '70-74', 12: '75-79', 13: '80+'}
bmi_labels = {1: 'Underweight', 2: 'Healthy', 3: 'Overweight', 4: 'Obese', 5: 'Severe Obesity'}

# Get unique physical activity levels
phys_activity_levels = df['PhysActivity'].unique()

# Define colors and markers for BMI categories
colors = sns.color_palette("tab10", 5)
markers = ['o', 's', '^', 'D', 'v']
for pa in phys_activity_levels:
    # Filter data for the current physical activity level
    pa_data = prevalence_data[prevalence_data['PhysActivity'] == pa]

    # Get sorted BMI categories
    bmi_categories = sorted(pa_data['BMICategory'].unique())

    # Create figure
    fig, ax = plt.subplots(figsize=(12, 8))

    # Plot lines for each BMI category
    for i, bmi in enumerate(bmi_categories):
        bmi_data = pa_data[pa_data['BMICategory'] == bmi]
        ax.plot(bmi_data['Age'], bmi_data['Diabetes_binary'], marker=markers[i], color=colors[i],
                label=bmi_labels.get(bmi, f'BMI {bmi}'))

    # Set x-axis labels with age group mappings
    ax.set_xticks(range(1, 14))
    ax.set_xticklabels([age_labels.get(i, str(i)) for i in range(1, 14)], rotation=45)

    # Set labels and title
    ax.set_xlabel('Age Group', fontsize=16)
    ax.set_ylabel('Diabetes Prevalence', fontsize=16)
    ax.set_title(f'Diabetes Prevalence by Age and BMI\nPhysical Activity: {"Yes" if pa == 1 else "No"}', fontsize=18, fontweight='bold')

    # Add legend
    ax.legend(title='BMI Category', fontsize=12)

    # Save the plot
    plt.tight_layout()
    plt.savefig(f'line_plot_pa_{pa}.png', dpi=300)
    plt.close()
```
